```python
import json
url= 'https://countriesnow.space/api/v0.1/countries/positions'
position = requests.get(url)

data = position.json()
 
# Open a json writer, and use the json.dumps()
# function to dump data
with open('/content/drive/MyDrive/Country_Position.json', 'w', encoding='utf-8') as jsonf:
  jsonf.write(json.dumps(data, indent=4))
```

```python
import csv
 
# Opening JSON file and loading the data
# into the variable data
with open('/content/drive/MyDrive/Country_Position.json') as json_file:
    data = json.load(json_file)
 
population = data['data']
 
# now we will open a file for writing
data_file = open('/content/drive/MyDrive/Country_Position.csv', 'w')
 
# create the csv writer object
csv_writer = csv.writer(data_file)
 
# Counter variable used for writing
# headers to the CSV file
count = 0
 
for country in population:
    if count == 0:
 
        # Writing headers of CSV file
        header = country.keys()
        csv_writer.writerow(header)
        count += 1
 
    # Writing data of CSV file
    csv_writer.writerow(country.values())
 
data_file.close()
```

```python
# if you do not have write access to this path, you will need 
# to change it to a path that you do have write access to.
gdrivePath="/content/drive/MyDrive/"
import sqlite3
import pandas as pd
import os
import pandasql
try:
  os.remove(gdrivePath+"Countries.db") # remove the database if it is already there
  print("removed earlier copy of Countries.db")
except:
  print("not already there")  
# create a connection to a database
conn = sqlite3.connect(gdrivePath+"Countries.db")
```

```python
cursor = conn.cursor()
 
# Table Definition
create_table = '''
CREATE TABLE "Country_Position" (
	"name"	TEXT,
	"iso2"	TEXT,
	"long"	REAL,
	"lat"	REAL,
	PRIMARY KEY("name"),
	FOREIGN KEY("iso2") REFERENCES "Country_Currency"("iso2"),
	FOREIGN KEY("name") REFERENCES "Country_Crime"("Country")
);
'''
 
# Creating the table into our
# database
cursor.execute(create_table)
 
# Opening the person-records.csv file
file0 = open('/content/drive/MyDrive/Country_Position.csv')
# Reading the contents of the
# person-records.csv file
contents = csv.reader(file0)
next(contents)
 
# SQL query to insert data into the
# person table
insert_records = "INSERT INTO Country_Position(name, iso2, long, lat) VALUES(?, ?, ?, ?)"
 
# Importing the contents of the file
# into our person table
cursor.executemany(insert_records, contents)
 
# SQL query to retrieve all data from
# the person table To verify that the
# data of the csv file has been successfully
# inserted into the table
select_all = "SELECT * FROM Country_Position"
rows = cursor.execute(select_all).fetchall()
 
# Output to the console screen
for r in rows:
    print(r)
 
# Committing the changes
conn.commit()
```

```python
import requests

url = 'https://query.wikidata.org/sparql'
query = '''
SELECT DISTINCT ?country ?countryLabel
{
  ?country wdt:P31 wd:Q6256 ;
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en"  }
}
'''
r = requests.get(url, params = {'format': 'json', 'query': query})
data = r.json()
```

```python
from google.colab import  drive
drive.mount('/content/drive')
```

```python
from pandas.core.frame import treat_as_nested
df = pd.json_normalize(data['results']['bindings'])
df = df.drop(columns=['country.type', 'countryLabel.type', 'countryLabel.xml:lang'], axis=1)
df = df.rename(columns={"countryLabel.value": "Country", "country.value": "Identifier"})
df['Identifier'] = df['Identifier'].str.replace('http?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+/', ' ', regex=True)
df = df.iloc[:,[1,0]]
df
```

```python
df.to_csv('/content/drive/MyDrive/SPARQL_Country_Identifier.csv',index=False, header=None) 
```
